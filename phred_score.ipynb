```python
from google.colab import drive
import pandas as pd

# 1. Mount Google Drive
drive.mount('/content/drive')

def parse_fastq(file_path, threshold=20):
    low_quality_records = []

    with open(file_path, "r") as f:
        while True:
            header = f.readline().strip()
            if not header:
                break
            sequence = f.readline().strip()
            plus = f.readline().strip()
            quality = f.readline().strip()

            phred_scores = [ord(ch) - 33 for ch in quality]  # Phred+33
            mean_score = sum(phred_scores) / len(phred_scores)

            if mean_score < threshold:
                low_quality_records.append({
                    "Read_ID": header,
                    "Sequence": sequence,
                    "Mean_Phred_Score": round(mean_score, 2)
                })

    return pd.DataFrame(low_quality_records)


# 2. Ask user for file path (inside Google Drive)
fastq_file = input("/content/drive/MyDrive/SRR34999326.fastq ").strip()
threshold = int(input("Enter Phred threshold (default 20): ") or 20)

# 3. Run analysis
result_df = parse_fastq(fastq_file, threshold)

# 4. Output
if result_df.empty:
    print("✅ No low-quality sequences found!")
else:
    print("⚠️ Low-quality sequences:")
    print(result_df.to_string(index=False))

    # Save results back to Google Drive
    output_path = "/content/drive/MyDrive/low_quality_sequences.csv"
    result_df.to_csv(output_path, index=False)
    print(f"\nResults saved to: {output_path}")
```
